```python
from graphviz import Digraph


def draw_heap(max_heap_array, item_count):
    dot = Digraph()
    # Create nodes
    for i in range(item_count):
        node = max_heap_array[i]
        # Display score and pair tuple
        label = f"{node.pair}\n{node.score}"
        dot.node(str(i), label)

        # Connect to parent
        if i > 0:
            parent_idx = (i - 1) // 2
            dot.edge(str(parent_idx), str(i))

    return dot
```
